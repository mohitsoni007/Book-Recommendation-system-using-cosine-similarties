```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('book.csv',encoding = "ISO-8859-1")
df
```

```python
df=df.rename({'User.ID':'userid','Book.Title':'name','Book.Rating':'rating'},axis=1)
df
```

```python
len(df.userid.unique())
```

```python
len(df.name.unique())
```

```python
User_book_data=df.pivot_table(index='userid',
                        columns='name',
                        values='rating').reset_index(drop=True)
```

```python
User_book_data
```

```python
User_book_data.index=df.userid.unique()
User_book_data
```

```python
User_book_data.fillna(0,inplace=True)
User_book_data
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( User_book_data.values,metric='cosine')
```

```python
user_sim
```

```python
user_sim_df=pd.DataFrame(user_sim)
```

```python
user_sim_df.index = df.userid.unique()
user_sim_df.columns = df.userid.unique()
```

```python
user_sim_df.iloc[0:10, 0:10]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:10, 0:10]
```
